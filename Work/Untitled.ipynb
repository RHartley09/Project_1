```python
data = pd.read_csv ('data/county_data_merged.csv')

Male = data['Total_Male'].mean()
Female = data['Total_Female'].mean()
fig1, ax1 = plt.subplots(figsize=(6, 5))
fig1.subplots_adjust(0.3,0,1,1)
my_data = [Male,Female]
my_labels = 'Male','Female'
plt.pie(my_data,labels=my_labels,autopct='%1.1f%%')
plt.title('Male vs Female Cases')
plt.axis('equal')
plt.savefig("MyGraphs/MaleFemale.png")
plt.show()
```

```python
data = pd.read_csv ('data/county_data_merged.csv')

ZeroToSeventeen = data['Total_age0to17'].mean()
EighteenTo64 = data['Total_age18to64'].mean()
SixtyFiveplus = data['Total_age65plus'].mean()
EightyFiveplus = data['Total_age85plusr'].mean()

fig1, ax1 = plt.subplots(figsize=(6, 5))
fig1.subplots_adjust(0.3,0,1,1)
my_data = [ZeroToSeventeen,EighteenTo64,SixtyFiveplus,EightyFiveplus]
my_labels = '0-17','18-64','65+','85+'
plt.pie(my_data,labels=my_labels,autopct='%1.1f%%')
plt.title('Age groups')
plt.axis('equal')
plt.savefig("MyGraphs/AgeGroups.png")
plt.show()
```
