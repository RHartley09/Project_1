```python
# sort by date
cov_counties2Redo_df = cov_counties_2_df.sort_values(by=['date'])
cov_counties2Redo_df
```

```python
# drop duplicate county entries to show latest numbers\
cov_counties2Redo_df = cov_counties2Redo_df.drop_duplicates(subset=['county', 'state'], keep="last")
```
